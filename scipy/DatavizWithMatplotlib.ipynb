## Introduction

Matplotlib is a Python package:

    - Based on NumPy

Three operation modes:

    - Scripts. Must use plt.show() to enter event loop. Use it once!
    - IPython shell. Must use %matplotlib
    - IPython notebook. Two modes
        * %matplotlib inline
        * %matplotlib notebook


## Initial setup

```python
import matplotlib.pyplot as plt

import numpy as np
```

Graphical device must by initialized to avoid unexpected behavior. Two modes:
    - notebook: interactive, only for browser-based devices
    - inline: static

```python
%matplotlib notebook

x = np.linspace(0, 10, 100) # We begin to do things
plt.plot(x, np.cos(x)) # And here we go!
```

    - inline: static

```python
%matplotlib inline

plt.plot(x, np.cos(x)) # And here we go!
```

## Matplotlib interfaces

Matplotlib comes with two different interfaces.

- Matlab-like. Old-fashioned function-oriented API.

```python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

- Object-oriented. Object-oriented and more powerfull API.

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

## Basic plots

Matplotlib requires a figure and axes. Plots are built on their top.

```python
plt.style.use('seaborn-whitegrid') # Setup theme

fig = plt.figure()
ax = plt.axes()
```

Then we can add the plot:

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x))
```

A basic Matplotlib script has the following form:

```python
%matplotlib

import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x))

plt.show()
```

Adjust lines

```python
%matplotlib inline

plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

Axis labeling

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

## Scatter plots

```python
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(np.random.random(5), np.random.random(5), marker, label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```

## Distribution plots

```python
plt.hist(np.random.randn(1000));
```

```python
plt.hist(np.random.randn(1000), bins=30, alpha=0.5);
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

plt.hist(x1, histtype='stepfilled', alpha=0.3, density=True, bins=40)
plt.hist(x2, histtype='stepfilled', alpha=0.3, density=True, bins=40)
plt.hist(x3, histtype='stepfilled', alpha=0.3, density=True, bins=40);
```

## Three-dimensional plots

Prepare data.

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

3-D visualizations:

```python
plt.contour(X, Y, Z, colors='black');
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```
