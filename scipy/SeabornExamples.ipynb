```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
sns.set()
```

```python
g = sns.factorplot("day", "total_bill", "sex", data=tips, kind="box")
g.set_axis_labels("Day", "Total Bill")
```

```python
with sns.axes_style(style='ticks'):
    g = sns.violinplot("day", "total_bill", "sex", data=tips)
```

```python
sns.violinplot("day", "total_bill", "sex", data=tips)
```

```python
sns.jointplot("total_bill", "tip", tips , kind="hex ")
```

```python
sns.set()
sns.jointplot("total_bill", "tip", tips , kind="hex")
```

```python
g = sns.pairplot(tips, vars=["tip", "total_bill"])
```

```python
sns.pairplot(iris , hue="species", palette="husl", markers=["o", "s", "D"], diag_kind='kde')
```

```python
ax = sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True)
```

```python
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)

plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True, col="smoke")
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(sns.distplot, "tip")
```

```python
g = sns.FacetGrid(tips, col="time", row="sex")
g.map(sns.distplot, "tip")
```
